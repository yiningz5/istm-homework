**homework1:is pandas an efficient way of handling big data? is it scale up?**

Pandas may not be a good choice in handling big data. Pandas run on a single machine and it has its own algorithm and memory constraints, which cannot load the data bigger than your memory allows and may be difficult when dealing big volume data.

Pandas scales up for some reasons. One is that when pandas represents data, it would uncompress the data to do faster processing, resulting in taking up bigger memory. Another reason is its way to save the data. For example, when pandas save a file with number 1, it would be recognized as integers rather than characters and takes up more bytes.

```
def pclass_name(x):
  if x == 1:
    x = '1st class'
  if x == 2:
    x = '2nd class'
  if x == 3:
    x = '3rd class'
  return x    
```
